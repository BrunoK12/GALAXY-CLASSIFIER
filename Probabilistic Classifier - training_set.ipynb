```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)  # Muestra todas las columnas
pd.set_option('display.expand_frame_repr', False)
df = pd.read_csv(r"C:\Users\PC\OneDrive\Escritorio\PUCP\2025-0\DEEP LEARNING\PROYECTO\BASE DE DATOS\TABLAS DE PROPORCIONES\training_solutions_rev1.csv")
df.head(5)
```

```python
df['prom_class11'] = df[['Class11.1', 'Class11.2', 'Class11.3', 'Class11.4', 'Class11.5', 'Class11.6']].mean(axis=1)
df['class'] = np.where(df['prom_class11']>0.1, 'espiral', '')
#print(df[df["class"] == "espiral"])

tot = df.shape[0]
esp = (df["class"] == "espiral").sum()
print("registros totales: ",tot)
print("registros 'espirales': ",esp)
```

```python
df['class'] = np.where(
    (df['Class7.1'] > 0) &  
    (df['Class7.1'] > df['Class7.2']) &  
    (df['Class7.1'] > df['Class7.3']) &  
    (df['class'] == ''),  
    'sphere',  
    df['class']
)

esfe = (df["class"] == "sphere").sum()
asign = esp + esfe

print("registros totales: ",tot)
print("registros 'espirales': ",esp)
print("registros 'sphere': ", esfe)
print("registros asignados: ", asign)
print("registros faltantes: ",tot-asign)
```

```python
# Class5.3 > (Class5.1  AND Class5.2) then class is "elliptical" then delete Class5.1, Class5.2, Class5.3
df['class'] = np.where(
    (df['Class5.3'] > 0) &  
    (df['Class5.3'] > df['Class5.1']) &  
    (df['Class5.3'] > df['Class5.2']) &  
    (df['class'] == ''),  
    'elliptical',  
    df['class']
)

elli = (df["class"] == "elliptical").sum()
asign = esp + esfe + elli

print("registros totales: ",tot)
print("registros 'espirales': ",esp)
print("registros 'sphere': ",esfe)
print("registros 'elliptical': ",elli)
print("registros asignados: ",asign)
print("registros faltantes: ",tot - asign)
```

```python
df['class'] = np.where(
    (df['Class6.1'] > df['Class6.2']) &  
    (df['class'] == ''),  
    'irregular',  
    df['class']
)

irre = (df["class"] == "irregular").sum()
asign = esp + esfe + elli + irre

print("registros totales: ",tot)
print("registros 'espirales': ",esp)
print("registros 'sphere': ",esfe)
print("registros 'elliptical': ",elli)
print("registros 'irregular': ",irre)
print("registros asignados: ",asign)
print("registros faltantes: ",tot - asign)
```

```python
columnas_a_conservar = ["GalaxyID", "class"]  
data = df[columnas_a_conservar] 
data
```
